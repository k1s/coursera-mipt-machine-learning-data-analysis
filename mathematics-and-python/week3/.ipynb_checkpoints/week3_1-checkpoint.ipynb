```python
from scipy import optimize
from numpy import sin, exp
from matplotlib import pyplot
%matplotlib inline
    
f = lambda x: sin(x / 5.) * exp(x / 10.) + 5. * exp(-x / 2.)

def step(xs):
    pyplot.plot(xrange(1, 30), [f(x) for x in xrange(1, 30)])
    pyplot.plot(xs, [f(x) for x in xs])
    
min = scipy.optimize.minimize()    
```

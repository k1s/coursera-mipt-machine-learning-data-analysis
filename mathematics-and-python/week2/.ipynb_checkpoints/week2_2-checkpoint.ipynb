```python
from scipy import linalg
from numpy import sin, exp
from matplotlib import pyplot
%matplotlib inline
    
f = lambda x: sin(x / 5.) * exp(x / 10.) + 5. * exp(-x / 2.)

rowA = lambda x: [x ** n for n in xrange(0, 2)]

def step(xs):
    A = [rowA(x) for x in xs]
    b = [f(x) for x in xs]
    solution = linalg.solve(A, b)
    pyplot.plot(xs, [f(x) for x in xs])
    pyplot.plot(xs, [x for x in solution])
```

```python
A = [rowA(x) for x in [1, 15]]
b = [f(1), f(15)]
solution = linalg.solve(A, b)
pyplot.plot(xrange(1, 15), [f(x) for x in xrange(1, 15)])
pyplot.plot([1, 15], [x for x in solution])
```

```python
A = [[1 ** n for n in xrange(0, 2)], [8 ** n for n in xrange(0, 2)], [15 ** n for n in xrange(0, 2)]]
b = [f(x) for x in [1, 8, 15]]
solution = linalg.solve(A, b)
pyplot.plot(xrange(1, 15), [f(x) for x in xrange(1, 15)])
pyplot.plot([1, 15], [x for x in solution])
```

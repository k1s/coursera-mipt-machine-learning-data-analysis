```python
from __future__ import print_function
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import numpy as np
import codecs
import re

magic = 42
```

```python
# возьмем имена номенклатур
mkString = lambda xs: ' '.join(xs)
words = lambda s: s.strip().replace('\\', '').split(" ")
withoutDigits = lambda s: re.sub('\d', '', s)
tokenize = lambda s: [withoutDigits(x) for x in words(s) if len(withoutDigits(x)) > 3]

content = codecs.open("nom3utf.csv", 'r', 'utf8')
corpus = [mkString(tokenize(x)) for x in content.readlines()]
```

```python
# превратим в чиселки
vectorizer = TfidfVectorizer(max_df=0.5, min_df=2)
X = vectorizer.fit_transform(corpus)
```

```python
# откластеризуем
km = KMeans(n_clusters=magic, init='k-means++', max_iter=100, n_init=1)
km.fit(X)
order_centroids = km.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
```

```python
# фиточай трава лек листья порош сбор
for i in xrange(0, magic):
    print('')
    for ind in order_centroids[i, :magic]:
        print(' %s' % terms[ind], end='')
    print()
```
